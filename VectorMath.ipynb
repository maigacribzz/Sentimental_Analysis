```python
from scipy import spatial
```

```python
cos_sim = lambda vec1,vec2: 1-spatial.distance.cosine(vec1,vec2)
```

```python
kingvec = nlp.vocab['king'].vector
manvec = nlp.vocab['man'].vector
womanvec = nlp.vocab['woman'].vector
```

```python
comp_sims = []
for word in nlp.vocab:
    if word.has_vector:
        if word.is_lower:
            if word.is_alpha:
                sim = cos_sim(kingvec,word.vector)
                comp_sims.append((word,sim))
```

```python
len(comp_sims)
```

```python
comp_sims = sorted(comp_sims,key = lambda item:-item[1])
```

```python
print([t[0].text for t in comp_sims[:10]])
```
